```python
import pandas as pd
import logging
import sys
from pathlib import Path
import json

from pythoneeg import visualization as viz
```

```python
data_folder = '/mnt/isilon/marsh_single_unit/PythonEEG/notebooks/tests/test-wars-sox5-3'
data_folder = Path(data_folder)

animal_ids = [p.name for p in data_folder.glob("*") if p.is_dir()]
df = []
for animal_id in animal_ids:
    json_files = list(data_folder.glob(f"{animal_id}/*.json"))
    json_data = json.load(open(json_files[0]))
    df.append({
        'id': json_data['animal_id'],
        'animal_name': animal_id,
        'genotype': json_data['genotype']
    })

df = pd.DataFrame(df)
df
```

```python
# Create a custom sorting order for genotypes
genotype_order = ['FWT', 'FHet', 'FMut', 'MWT', 'MHet', 'MMut']

# Sort the dataframe using the custom order
df = df.assign(genotype_order=df['genotype'].map({g: i for i, g in enumerate(genotype_order)}))
df = df.sort_values('genotype_order').drop('genotype_order', axis=1)
df
```
